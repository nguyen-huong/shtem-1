```python
import numpy as np
import matplotlib.pyplot as plt
import csv
import pandas as pd
from PIL import Image
```

## Data Visualization

### This section covers parks data.


```python
data = pd.read_csv('parks.csv')
data
```

```python
print(np.array(data[['Name']]))
```

```python
parks = ['names']
with open('parks.csv') as a:
    reader = csv.reader(a, delimiter=',')
    count = 0
    for row in reader:
        parks.append(str(row[0]))
        count = count + 1
        
print(parks)
```

```python
x = np.array(data[['Longitude']])
y = np.array(data[['Latitude']])

plt.figure()
plt.scatter(x,y)
plt.title('Park Location')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

```python
labels = np.array(data[['Name']])

fig, ax = plt.subplots()
ax.scatter(x,y)

for i, txt in enumerate(labels):
    ax.annotate(txt,(x[i],y[i]))
```
